```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = np.load("../Data/KNN/Regression/X_data.npy"), np.load("../Data/KNN/Regression/Y_data.npy")
```

```python
plt.scatter(X, y)
plt.show()
```

```python
m = 30
c = 0

y_pred = m * X + c
```

```python
plt.plot(X, y_pred)
plt.scatter(X, y)
plt.show()
```

```python
xy_bar = (X * y.reshape(-1,1)).mean()
x_bar = X.mean()
y_bar = y.mean()
x_2_bar = (X**2).mean()
```

```python
m = (xy_bar - y_bar*x_bar)/(x_2_bar - (x_bar)**2)
c = y_bar - m*x_bar
```

```python
m, c
```

```python
plt.plot(X, m*X+c, color="red")
plt.scatter(X, y)
plt.show()
```
