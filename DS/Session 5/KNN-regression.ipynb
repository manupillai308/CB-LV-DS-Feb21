```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

X, y = np.load("../Data/KNN/Regression/X_data.npy"), np.load("../Data/KNN/Regression/Y_data.npy")
```

```python
X.shape, y.shape
```

```python
plt.scatter(X, y)
plt.title("Feature vs Output")
plt.show()
```

```python
def predict(test_point, k = 10):
    distances = np.linalg.norm(X - test_point, ord=2, axis=1)
    top_indexes = np.argsort(distances)[:k]

    y_pred = np.mean(y[top_indexes])
    return y_pred, top_indexes
```

```python
test_point = np.array([[2]])
# test_point.shape
plt.scatter(X, y)
y_pred, closest = predict(test_point)
plt.scatter(X[closest], y[closest], color="yellow")
plt.scatter(2, y_pred, color="red")
plt.title("Feature vs Output")
plt.show()
```

```python

# test_point.shape
plt.scatter(X, y)
pred = []
for i in np.arange(-3.0, 3.1, 0.01):
    test_point = np.array([[i]])
    y_pred, closest = predict(test_point)
    # plt.scatter(i, y_pred, color="red")
    pred.append(y_pred)
plt.plot(np.arange(-3.0, 3.1, 0.01), pred, color="red")
plt.title("Feature vs Output")
plt.show()
```
